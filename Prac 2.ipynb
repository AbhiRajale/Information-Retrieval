Levenshtein Distance

Find for: 1. Price and Rice
          2. Elephant and Relevant
          3. Intention and Extension

```python
#Initilizing DataFrame
import pandas as pd

word1 = list("0" + input("Enter First Word: "))
word2 = list("0" + input("Enter Second Word: "))

df = pd.DataFrame(data = 0, index = word1, columns = word2)
df
```

```python
#Adding Starting Values
for i in range(len(word1)):
    for j in range(len(word2)):
        if i == 0 or j == 0:
            df.iloc[i, j] = i + j
        else:
            break
df
```

```python
#Finding Minimum Operations
for i in range(1, len(word1)):
    for j in range(1, len(word2)):
        l = [df.iloc[i - 1, j] + 1, df.iloc[i, j - 1] + 1]
        if word1[i] == word2[j]:
            l.append(df.iloc[i - 1, j - 1])
        else:
            l.append(df.iloc[i - 1, j - 1] + 1)
        df.iloc[i, j] = min(l)

print("Minimum Operations Required:", df.iloc[len(word1) - 1, len(word2) - 1])
df
```

Outputs:
    1. Minimum Operations Required: 1
    2. Minimum Operations Required: 3
    3. Minimum Operations Required: 3
